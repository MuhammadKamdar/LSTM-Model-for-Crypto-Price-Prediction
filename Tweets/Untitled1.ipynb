```python
import pandas as pd
import numpy as np
import pickle as pkl
```

```python
with open("chartdata.pkl", "rb") as f:
    object = pkl.load(f)
df = pd.DataFrame(object)
```

```python
chunk = pd.read_csv('btc_coinMarket.csv', chunksize=100000, lineterminator='\n', low_memory=False)
df = pd.concat(chunk)
```

```python
df.head()
```

```python
df[['open', 'high', 'low','close', 'volume', 'marketCap','timestamp']] = df['open;high;low;close;volume;marketCap;timestamp\r'].str.split(';', expand=True)
df.drop('open;high;low;close;volume;marketCap;timestamp\r', axis=1, inplace=True)
```

```python
df.head()
```

```python
from dateutil.parser import parse
```

```python
df['timestamp'] = df['timestamp'].apply(lambda x: parse(x))

# Extract date from datetime and assign it to a new column
df['Date'] = df['timestamp'].dt.date
```
